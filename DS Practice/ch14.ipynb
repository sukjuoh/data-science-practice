```python
formula = "price ~ " + " + ".join(df_sub.columns[:-1])
y, X = dmatrices(formula, df_sub, return_type = "dataframe")

df_vif = pd.DataFrame()
df_vif["colnames"] = X.columns
df_vif["VIF"] = [vif(X.values, i) for i in range(X.shape[1])]
df_vif
```

#2 Price를 종속변수로 하고 carat과 depth를 독립변수로 하여 생성한 선형 회귀 모델을 사용하여 알아본 carat이 1이고, depth가 60, table이 55인 다이아몬드의 가격은 얼마인가?

```python
from statsmodels.formula.api import ols
model = ols("price ~ carat + depth", data = df).fit()
model.summary()
```

```python
model.predict({"carat" : 1, "depth" : 60, "table" : 55})
```

#3 price를 종속변수로 하고 carat, color, depth를 독립변수로 하여 생성한 선형 회귀 모델을 사용하여 알아본 carat이 1이고, depth가 50, color가 E인 다이아몬드 가격은 얼마인가?

```python
df_dum = pd.get_dummies(df, columns = ["color"], drop_first = True)
df_dum.head(2)
```

```python
formula = "price ~ carat + depth + color_E + color_F + color_G + color_H + color_I + color_J"
model = ols(formula, data = df_dum).fit()
model.summary()
```

```python
model.predict({"carat" : 1, "depth" : 50, "color_E" : 1,
                "color_F" :0, "color_G" : 0, "color_H" : 0, "color_I" : 0, "color_J" : 0})
```
